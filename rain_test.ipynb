```python
import pandas as pd
import os
```

```python
avg_rain_data = os.path.join('Resources/avg_rain_state.csv')
rain_df = pd.read_csv(avg_rain_data)
```

```python
rain_df.columns
```

```python
rain_df.sort_values(['Rank'], ascending=False).tail()
```

```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_axis = np.arange(len(rain_df))
tick_locations = [value for value in x_axis]
```

```python
# Create graph
tick_locations = [value for value in x_axis]
plt.figure(figsize=(20,3))
plt.bar(x_axis, rain_df["Inches"])
plt.xticks(tick_locations, rain_df["State"], rotation='vertical')
plt.title("Rain in Inches by State")
plt.xlabel("State")
plt.ylabel("Rain (inches)")
plt.show()
```

```python
rain_df['Inches'].mode()
```

```python
state_and_inches = rain_df[["State","Inches"]]

# Set the index to be "State" so they will be used as labels
state_and_inches = state_and_inches.set_index("State")

state_and_inches.mode()
```
